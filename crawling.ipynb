```python
# selenium으로 하기 ###############################
for game in sub:
    # 각 경기 결과 페이지에 들어가기
    driver.get(game)
    time.sleep(5)
    
    # 해당 페이지에서 통계 클릭
    driver.find_element_by_xpath("""//*[@id="a-match-statistics"]""" ).click()
    time.sleep(3)
    
    # html 가져온 후, BeautifulSoup 객체 만들기.
    html2 = driver.page_source
    soup2 = BeautifulSoup(html2, 'html.parser')

    
    # 팀 이름 담겨있는 html
    name_list = soup2.find_all("img", {"title" : "프로필 보기"})

    # 스코어 담겨있는 html
    score_info = soup2.find_all("span", {"class" : "scoreboard"})

    home_score = int(score_info[0].get_text()) # 홈팀 스코어
    away_score = int(score_info[1].get_text()) # 어웨이팀 스코어
    
    # 해당 통계 페이지 html 가져오기
    stats = soup2.find("div", {"id" : "tab-statistics-0-statistic"}).find_all("div", {"class" : "statRow"})

    
    ######### 홈팀 가져오기 #########
    
    # 홈팀이름 가져오기
    team_name.append(name_list[0]['alt'])
    
    # 스코어 가져오기
    score.append(home_score)

    if home_score > away_score:
        result.append(2)
    elif home_score == away_score:
        result.append(1)
    else :
        result.append(0)

    # 날짜 가져오기
    date.append(soup2.find("div", {"id" : "utime"}).get_text())
        
    # 통계 가져오기
    for i in range(len(stats)):
        home_stats = stats[i].find("div", {"class" : "statText statText--homeValue"}).get_text().strip().replace("%","")
        #away_stats = stats[i].find("div", {"class" : "statText statText--awayValue"}).get_text().strip().replace("%","")
        # 변수이름 가져옴
        byeonsu = stats[i].find("div", {"class" : "statText statText--titleValue"}).get_text().strip().replace("%","")
        matching(byeonsu, home_stats)
    
    ######### 어웨이팀 가져오기 #########
    
    # 팀 이름 가져오기
    team_name.append(name_list[1]['alt'])
    
    # 스코어 가져오기
    score.append(away_score)
    
    if home_score > away_score:
        result.append(0)
    elif home_score == away_score:
        result.append(1)
    else :
        result.append(2)
        
    # 날짜 가져오기
    date.append(soup2.find("div", {"id" : "utime"}).get_text())
        
    # 통계 가져오기
    for i in range(len(stats)):
        #home_stats = stats[i].find("div", {"class" : "statText statText--homeValue"}).get_text().strip().replace("%","")
        away_stats = stats[i].find("div", {"class" : "statText statText--awayValue"}).get_text().strip().replace("%","")
        # 변수이름 가져옴
        byeonsu2 = stats[i].find("div", {"class" : "statText statText--titleValue"}).get_text().strip().replace("%","")
        matching(byeonsu2, away_stats)
    
```
