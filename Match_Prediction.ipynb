# 데이터 전처리

#### 시계열 형태로 변환

```python
df['date'] = pd.to_datetime(df['date'], format='%d.%m.%Y %H:%M')
df.info()
```

```python
df.head()
```

```python
df.set_index('date', inplace = True)
```

```python
df.head()
```

team_name 명목형 변수를 제거하는 대신 전 시즌의 리그 순위를 이용하여 수치형 변수로 변환

```python
df['team_name'].value_counts()
```

```python
def ranking(name):
    if name == '맨체스터 시티':
        return 1
    elif name == '리버풀':
        return 2
    elif name == '첼시':
        return 3
    elif name == '토트넘':
        return 4
    elif name == '아스날':
        return 5
    elif name == '맨체스터 유나이티드':
        return 6
    elif name == '울버햄튼':
        return 7
    elif name == '에버튼':
        return 8
    elif name == '레스터':
        return 9
    elif name == '웨스트햄':
        return 10
    elif name == '왓포드':
        return 11
    elif name == '크리스탈팰리스':
        return 12
    elif name == '뉴캐슬 Utd':
        return 13
    elif name == '본머스':
        return 14
    elif name == '번리':
        return 15
    elif name == '사우샘프턴':
        return 16
    elif name == '브라이튼':
        return 17
    else:
        return 18
```

```python
df['last_seson_ranking'] = df['team_name'].apply(ranking)
df.head()
```

#### 데이터 분할

```python
df_x = df.drop(columns = ['result','team_name'], axis=1)
df_x.head()
```

```python
df_y = df.pop('result')
df_y.head()
```

훈련, 평가 데이터 분할

```python
from sklearn.model_selection import train_test_split
```

```python
train_x, test_x, train_y, test_y = train_test_split(df_x, df_y, test_size=0.33, random_state=42)
```

# 모델링

RandomForest 

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
RF_model = RandomForestClassifier()
```

```python
RF_fitted = RF_model.fit(train_x, train_y)
```

```python
predicted_y = RF_fitted.predict(test_x)
predicted_y
```

확률값으로 확인하기

```python
RF_fitted.predict_proba(test_x)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(test_y, predicted_y)
```
